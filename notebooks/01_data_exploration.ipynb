```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data_dir = "../data"

users = pd.read_csv(os.path.join(data_dir, "users.csv"))
products = pd.read_csv(os.path.join(data_dir, "products.csv"))
ratings = pd.read_csv(os.path.join(data_dir, "ratings.csv"))

print(f"Пользователи: {users.shape}")
print(f"Продукты: {products.shape}")
print(f"Рейтинги: {ratings.shape}")
```

### Примеры данных

```python
display(users.head())
display(products.head())
display(ratings.head())
```

### Возраст и доход пользователей

```python
users[['age', 'income']].hist(figsize=(10,4))
plt.suptitle("Распределение пользователей по возрасту и доходу")
plt.show()
```

### Распределение рейтингов

```python
ratings['rating'].value_counts().sort_index().plot(kind='bar', title='Распределение рейтингов')
plt.show()
```

### Частота типов продуктов

```python
products['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Типы финансовых продуктов')
plt.show()
```
